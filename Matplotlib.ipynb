```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
# Functional way of plotting

x = np.linspace(0,5,11)
y = x**2
plt.title("Square Numbers", fontsize=24)
plt.xlabel("Value", fontsize=14)
plt.ylabel("Square Value", fontsize=14)
plt.plot(x,y)
```

```python
# Functional way of plotting

plt.subplot(1,2,1)
plt.plot(x,y,"r")

plt.subplot(1,2,2)
plt.plot(y,x,"b")
```

```python
# Object Oriented Method 

fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Title')
axes.plot(x,y)
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.15,0.55,0.38,0.29])
axes1.plot(x,y)
axes2.plot(x,y,'purple')
```
